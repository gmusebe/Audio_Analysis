```python
# Initialize Our Model
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
data_N=data[data['Message'].isnull()==False].reset_index().drop('index', axis=1)
```

```python
Sentences=data_N['Message'].to_list()
```

```python
sentence_embeddings = model.encode(Sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
for i in range(data_N.shape[0]):
    L=cosine_similarity([sentence_embeddings[i]],np.delete(sentence_embeddings, i, 0))
    T=L.tolist()[0]
    K=[e for e, j in enumerate(T) if j == max(T)]
    data_N.loc[i, 'Score']=max(T)
    K_1=[e for e in K if e >=i ]
    ls_1=[elem for elem in K]
    ls=' '.join([str(elem) for elem in K])
    for j in ls_1:
        if j in K_1:
            ls=ls.replace(str(j), str(j+1))
    data_N.loc[i, 'index']=ls
```

```python
#Round the Score
data_N['Score']=round(data_N['Score'], 2)
```

```python
#Extract Similar posts Or quasi similar
data_N[data_N['Score']==1.00]
#data_N[data_N['Score']>=0.85]
```
